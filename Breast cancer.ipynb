```python
x=df.iloc[:,1:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
def mymodel(model):
    model.fit(xtrain, ytrain)
    ypred = model.predict(xtest)
    ac= accuracy_score(ytest, ypred)
    cm = confusion_matrix(ytest, ypred)
    cr = classification_report(ytest, ypred)
    print(f"Accuracy-: {ac}\n\nConfusion Matrix-: \n{cm}\n\nClassification Report-: \n{cr}")
```

```python
knn = KNeighborsClassifier()
logreg = LogisticRegression()
svm = SVC()
```

```python
mymodel(knn)
```

```python
mymodel(logreg)
```

```python
mymodel(svm)
```

- knn -> n_neighbors, metrics
- logreg -> C, penalty, solver
- svm -> kernel, C, gamma

# GridSearchCV

```python
para = {'C':[0.1, 1, 10, 100, 1000], 'gamma':[1, 0.1, 0.01, 0.001, 0.0001], 'kernel':["rbf"]}
```

```python
from sklearn.model_selection import GridSearchCV

grid  = GridSearchCV(SVC(), para, refit=True, verbose=3)

grid.fit(xtrain, ytrain)
```

```python
grid.best_params_
```

```python
gpred=grid.predict(xtest)
```

```python
print(accuracy_score(ytest,gpred))
print(classification_report(ytest,gpred))
```

```python
svm=SVC(C=1000, gamma=0.1, kernel="rbf")
```

```python
mymodel(svm)
```
